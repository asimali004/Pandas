```python
import pandas as pd
```

```python
#attributes of the dropna function for missing values
#axis=0                   1
#how=any                  all
#thresh=None
#subset=None               ["Col_name"]
#inplace=False            True
```

```python
df=pd.read_csv("Data.csv",names=["Names","S1","S2","S3","S4","S5","Total","Average","Grades"])
```

```python
df
```

```python
df.dropna()
```

```python
df.dropna(axis=1)#by default axis=0
```

```python
df.dropna(how="any")#by default
```

```python
df.dropna(how="all")#any is by default
#drops: if all values in a row or column are null
```

```python
df.dropna(how="any",axis=1)
```

```python
df.dropna(thresh=4)#should thresh NON-NULL values :: otherwise the axes will be skipped
```

```python
df.dropna(thresh=8)
```

```python
df.dropna(subset=["S5"])
```

```python
df.dropna(subset=["S4"])
```

```python
df.dropna(subset=["S5"],axis=1)
```

```python
df.dropna(subset=["S5"],axis=0)
```

```python
df.dropna(inplace=True)#create a new dataframe
```

```python
df
```

```python
df=pd.read_csv("Data.csv",names=["Names","S1","S2","S3","S4","S5","Total","Average","Grades"])
```

```python
df
```

```python
df.fillna("Hello")# 0/5/9.4/-3.33/"ali"
```

```python
df.fillna({"S2":"hy","Grades":"F"})#column name: value
```

```python
df.fillna(method="ffill")
```

```python
df.fillna(method="pad")
```

```python
df.fillna(method="bfill")
```

```python
df.fillna(method="backfill")
```

```python
df.fillna(axis=0)
```

```python
df.fillna(axis=0,method="ffill")
```

```python
df.fillna(axis=1,method="ffill")#note the output col 1
```

```python
df.fillna(axis=1,method="bfill")#note the output col -1
```

```python
df.fillna(5,limit=1)
```

```python
df.fillna(method="ffill",limit=1)
```

```python
df.fillna(inplace=True)
```

```python
df.fillna(method="ffill",limit=1,inplace=True)
df
```
